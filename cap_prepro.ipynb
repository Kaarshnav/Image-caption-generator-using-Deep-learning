```
img_maps_captlist={}
for line in captions:
  first,cap=line.split('\t')
  imgid=first.split(".")[0]
  if img_maps_captlist.get(imgid) is None:
    img_maps_captlist[imgid]=[]
  img_maps_captlist[imgid].append(cap.split("\n")[0])

  
```

```
import re
def dataclean(line):

  line=line.lower()
  line=re.sub("[^a-z]+"," ",line)
  return line


```
